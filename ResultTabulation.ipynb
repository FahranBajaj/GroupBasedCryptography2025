```python
import pandas as pd
```

```python
results = pd.read_csv("./CMC_FOLDER/AttackResults2.csv")
knownGroups = {
    "UnivGrigorchuk" : 6,
    "GS3" : 3, 
    "Hanoi3" : 3,
    "GRIGORCHUK" : 2,
    "IMG" : 2
}
def idToDegree(groupID):
    if groupID in knownGroups:
        return knownGroups[groupID]
    splitId = groupID.split("_")
    if splitId[1] == "STAB":
        return int(splitId[2])
    return int(splitId[1])
    
results["Degree"] = results["groupID"].map(lambda x : idToDegree(x))


def groupType(groupID):
    if groupID in knownGroups:
        return "known"
    splitId = groupID.split("_")
    if splitId[1] == "STAB":
        return "randomUsedForStabilization"
    if splitId[0] == "GGS":
        return "randomGGS"
    return "random"

results["Type"] = results["groupID"].map(lambda x : groupType(x))
results["Successful"] = results["outcome"].map(lambda x : int(x == 1))
results.head(10)
```

```python
print(results["groupID"].unique())
```

```python
def tabulate(degree, typeOfGroup):
    data = results[results["Degree"] == degree]
    data = data[data["Type"] == typeOfGroup]
    print(data.groupby(["list_size", "g_len", "r_len"]).agg({"Successful" : [sum, len], "time_attack" : "mean"}))

tabulate(20, "random")
```
